```python
import pandas as pd
import sqlite3
import datetime
# Create connection to database
con = sqlite3.connect("dublinBikes.db")
cur = con.cursor()

# Select all data
cur.execute("SELECT * From dublinBikes")

# Store all fetched data in a list
data = cur.fetchall()
con.close()
```

```python
# Create a pandas dataframe from this list
df = pd.DataFrame(data)
df.head(5)
```

```python
# Rename the column headers for clarity
df.columns=["address","available_bike_stands","available_bikes","banking","bike_stands","bonus","contract_name",\
            "last_update","name","number","position_lat","position_lng","status"]
df.head(10)
```

```python
# Convert to seconds
df["new_time"] = df["last_update"]/1000
```

```python
import requests
stamen = folium.Map(location=[53.343368, -6.270120], tiles='Stamen Toner', zoom_start=14)

r = requests.get('http://api.citybik.es/dublinbikes.json')
data = r.json()

for i in data:
    if (i['free'] > (i['bikes'] /3.0)):
        x = float(i['lat']/1000000)
        y = float(i['lng']/1000000)
        stamen.simple_marker([x, y], popup=i['name'],marker_color='green')
    else:
        x = float(i['lat']/1000000)
        y = float(i['lng']/1000000)
        stamen.simple_marker([x, y], popup=i['name'],marker_color='red')
stamen
```
